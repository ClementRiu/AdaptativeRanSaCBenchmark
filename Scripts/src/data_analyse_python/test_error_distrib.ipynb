```python
def readMagsac(path):
    with open(path, "r") as file:
        line = file.readline()
        line = np.array([float(value) for value in line.split(", ")[:-1]])
    return line
```

```python
def readMatches(path):
    data = []
    with open(path, "r") as file:
        for line in file:
            data.append(np.array([float(value) for value in line.split()]))
    return np.array(data)
    
```
